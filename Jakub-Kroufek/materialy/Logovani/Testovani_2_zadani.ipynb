# 2. úloha Závody

Zadání:

1. Vytvořte třídu Běžec, která bude reprezentovat běžce v závodu. Třída by měla obsahovat:
    - Jméno běžce
    - Věk běžce
    - Osobní rekord na 5 km (v minutách)
    - Metody pro získání a nastavení těchto atributů

2. Vytvořte třídu Závod, která bude reprezentovat běžecký závod. Třída by měla obsahovat:
    - Název závodu
    - Datum konání
    - Seznam běžců (objekty třídy Běžec)
    - Metody pro přidání a odebrání běžce ze závodu
    - Metodu pro vypsání všech běžců v závodu

3. Vytvořte automatické testy pro třídy Běžec a Závod:
    - Otestujte správné vytvoření objektů Běžec a Závod
    - Otestujte přidání a odebrání běžce ze závodu
    - Otestujte správné nastavení a získání atributů běžce
    - Otestujte výpis všech běžců v závodu

```python
import unittest
class Zavod():
    def __init__(self, nazev, datum):
        self.nazev = nazev
        self.datum = datum
        self.seznam = []
    def add_bezec(self, bezec):
        if bezec not in self.seznam:
            self.seznam.append(bezec)
        else:
            print("Běžec se již v seznamu nachází")
    def remove_bezec(self, bezec):
        if bezec in self.seznam:
            self.seznam.pop(self.seznam.index(bezec))
        else:
            print("Běžec se v seznamu nenachází")
    def get_bezci(self):
        temp = []
        for bezec in self.seznam:
            temp.append(f"{bezec}")
        return "".join(temp)
    def get_seznam(self):
        return self.seznam
    def get_nazev(self):
        return self.nazev
    def get_datum(self):
        return self.datum
class Bezec():
    def __init__(self, jmeno, vek, rekord=int):
        self.jmeno = jmeno
        self.vek = vek
        self.rekord = rekord
    def get_jmeno(self):
        return self.jmeno
    def get_vek(self):
        return self.vek
    def get_rekord(self):
        return self.rekord
    def set_jmeno(self, jmeno):
        self.jmeno = jmeno
    def set_vek(self, vek):
        self.vek = vek
    def set_rekord(self, rekord=int):
        self.rekord = rekord
    def __str__(self):
        return f"{self.jmeno}, {self.vek}, {self.rekord}\n"

class TestZavodaBezec(unittest.TestCase):
    def test_bezec(self):
        test_jmeno = "Adam"
        test_vek = 19
        test_rekord = 5
        Adam = Bezec(test_jmeno, test_vek, test_rekord)
        self.assertEqual(Adam.get_jmeno(), test_jmeno)
        self.assertEqual(Adam.get_vek(), test_vek)
        self.assertEqual(Adam.get_rekord(), test_rekord)
    def test_nastaveni_bezec(self):
        Subjekt = Bezec("Adam", 19, 3)
        jmeno_nastav = "Pepa"
        vek_nastav = 43
        rekord_nastav = 6
        Subjekt.set_jmeno(jmeno_nastav)
        Subjekt.set_rekord(rekord_nastav)
        Subjekt.set_vek(vek_nastav)
        self.assertEqual(Subjekt.get_jmeno(), jmeno_nastav)
        self.assertEqual(Subjekt.get_vek(), vek_nastav)
        self.assertEqual(Subjekt.get_rekord(), rekord_nastav)        
    def test_zavod(self):
        test_jmeno = "Prvni"
        test_datum = "10.4"
        turnaj = Zavod(test_jmeno, test_datum)
        self.assertEqual(turnaj.get_nazev(), test_jmeno)
        self.assertEqual(turnaj.get_datum(), test_datum)
    def test_pridani(self):
        Test_bezec = Bezec("Jmeno", 20, 4)
        Test_turnaj = Zavod("ExtraLiga", "10.4")
        Test_turnaj.add_bezec(Test_bezec)
        self.assertEqual(Test_turnaj.get_bezci(), Test_bezec.__str__())
    def test_odebrani(self):
        Test_bezec = Bezec("Jmeno", 20, 4)
        Test_turnaj = Zavod("ExtraLiga", "10.4")
        Test_turnaj.add_bezec(Test_bezec)
        Test_turnaj.remove_bezec(Test_bezec)
        Test_turnaj.get_bezci()
        self.assertEqual(Test_turnaj.get_bezci(), "")
    def test_vicero_bezcu(self):
        Test_bezec2 = Bezec("Jmeno", 20, 4)
        Test_bezec3 = Bezec("Name", 43, 6)
        Test_bezec4 = Bezec("Pepa", 19, 2)
        Test_turnaj2 = Zavod("ExtraLiga", "10.4")
        Test_turnaj2.add_bezec(Test_bezec2)
        Test_turnaj2.add_bezec(Test_bezec3)
        Test_turnaj2.add_bezec(Test_bezec4)
        result = f"{Test_bezec2}{Test_bezec3}{Test_bezec4}"
        self.assertEqual(Test_turnaj2.get_bezci(), result)        
        
#při testování běží definice nahoře dle abecedního pořadí. Je nutné při každém individuálním testu definovat nové objetky a hodnoty pro izolované a správné otestování
if __name__ == '__main__':
    unittest.main()
```

# 3. úloha letec


- Navrhněte tři třídy:

    - **Letec**
        - jméno
        - příjmení
        - počet nalétaných hodin
        - metody pro nastavení a získání všech údajů

    - **Letadlo**
        - název letadla
        - typ letadla
        - minimální požadovaný počet hodin pilota
        - metody pro nastavení a získání všech údajů

    - **Letiště**
        - počet drah
        - počet volných drah
        - rozpis příletů/odletů:
            - letadlo
            - letec
            - typ akce (přílet/odlet)
        - metody pro:
            - kontrolu volné dráhy
            - přidání příletu/odletu
            - ověření, zda pilot splňuje požadavek na počet hodin

- Zadání testů:
    - Ověřte, že pokud je dráha volná, může letec přistát nebo odletět.
    - Ověřte, že pilot bez dostatečného počtu hodin nemůže letět daným letadlem.
    - Ověřte správné zaznamenání příletu/odletu do rozpisu.

```python
class Letec():
    def __init__(self, jmeno, prijmeni, hodiny):
        self.jmeno = jmeno
        self.hodiny = hodiny
        self.prijmeni = prijmeni
    def get_jmeno(self):
        return self.jmeno
    def set_jmeno(self, jmeno):
        self.jmeno = jmeno
    def set_hodiny(self, hodiny):
        self.hodiny = hodiny
    def get_hodiny(self):
        return self.hodiny
    def get_prijmeni(self):
        return self.prijmeni
    def set_prijmeni(self, prijmeni):
        self.prijmeni = prijmeni
    def __str__(self):
        return f"{self.jmeno}, {self.hodiny}, {self.prijmeni}"
    
class Letadlo():
    def __init__(self, nazev, typ, min_hodiny):
        self.nazev = nazev
        self.typ = typ
        self.min_hodiny = min_hodiny
    def get_nazev(self):
        return self.nazev
    def get_typ(self):
        return self.typ
    def get_min_hodiny(self):
        return self.min_hodiny
    def set_nazev(self, nazev):
        self.nazev = nazev
    def set_typ(self, typ):
        self.typ = typ
    def set_min_hodiny(self, min_hodiny):
        self.min_hodiny = min_hodiny
    def __str__(self):
        return f"{self.nazev}, {self.typ}, {self.min_hodiny}"            
    

class Letiste():
    def __init__(self, pocet_drah, pocet_volnych_drah):
        self.pocet_drah = pocet_drah
        self.pocet_volnych_drah = pocet_volnych_drah
        self.rozpis = []
    def add_odlet(self, letadlo, letec):
        if letadlo.get_min_hodiny() > letec.get_hodiny():
            return "Letec nemá dostatek hodin"
        for obsah in self.rozpis:
            if letadlo == obsah[0]:
                return "Letadlo je již v seznamu zahrnuto"
            elif letec == obsah[1]:
                return "Letec je již v seznamu zahrnutý"
        if self.kontrola_draha() == True:
            self.pocet_volnych_drah -= 1
            self.rozpis.append([letadlo, letec, "odleta"])
            return "zapsano"
    def add_prilet(self, letadlo, letec):
        if letadlo.get_min_hodiny() > letec.get_hodiny():
            return "Letec nemá dostatek hodin"
        for obsah in self.rozpis:
            if letadlo == obsah[0]:
                return "Letadlo je již v seznamu zahrnuto"
            elif letec == obsah[1]:
                return "Letec je již v seznamu zahrnutý"
        if self.kontrola_draha() == True:
            self.pocet_volnych_drah -= 1
            self.rozpis.append([letadlo, letec, "prileta"])
            return "zapsano"
    def get_seznam(self):
        temp = []
        for let in self.rozpis:
            temp.append(f"|{let[1].__str__()}| {let[2]} s |{let[0].__str__()}|")
        return "".join(temp)
    def kontrola_draha(self):
        if self.pocet_volnych_drah > 0:
            return True
        return False

import unittest
class TestTeplomeru(unittest.TestCase):
    def test_volna_draha(self):
        airbus = Letadlo("Airbus", "osobni", 400)
        subjekt = Letec("Subjekt", "Prijmeni", 401)
        praha = Letiste(19, 6)
        praha.add_odlet(airbus, subjekt)
        self.assertEqual(praha.get_seznam(), f"|{subjekt.__str__()}| odleta s |{airbus.__str__()}|")
    def test_nedostatek_hodin(self):
        airbus = Letadlo("Airbus", "osobni", 400)
        subjekt = Letec("Subjekt", "Prijmeni", 399)
        praha = Letiste(19, 6)
        self.assertEqual(praha.add_odlet(airbus, subjekt), "Letec nemá dostatek hodin")  
#při testování běží definice nahoře dle abecedního pořadí. Je nutné při každém individuálním testu definovat nové objetky a hodnoty pro izolované a správné otestování
if __name__ == '__main__':
    unittest.main()
```
