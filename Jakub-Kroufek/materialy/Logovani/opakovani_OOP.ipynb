Vytvořte třídu Knihovna, která obsahuje tyto atributy: (1b)
* nazev
* seznamKnih
* kapacitaKnih

Vytvořte tyto metody: (1b + 2b)

* get a set pro všechny atributy
* metodu pro přepis všech knih do stringu
* ToString
* přidání a odebrání knihy z knihovny (lze přidat a odebrat pouze knihy, které se tam vejdou nebo v knihovně ještě nejsou + zpětná vazba uživateli) (+2b)


```python
class Knihovna():
  def __init__(self, nazev, seznamKnih:list, kapacitaKnih:int):
    self.nazev = nazev
    self.seznamKnih = seznamKnih
    self.kapacitaKnih = kapacitaKnih
  def getNazev(self):
    print(self.nazev)
  def getSeznamKnih(self):
    temp = []
    for kniha in self.seznamKnih:
      temp.append(f"{kniha.getNazev()}, {kniha.getAutor()}, {kniha.getZanr()}, {kniha.getPocetStran()}")
    return(f"| {' | '.join(temp)} |")
  def getKapacitaKnih(self):
    print(self.kapacitaKnih)
  def addKniha(self, kniha):
    if kniha not in self.seznamKnih:
      if self.kapacitaKnih <= len(self.seznamKnih):
        print("Kniha se do knihovny nevejde")
      else:
        self.seznamKnih.append(kniha)
        print("Kniha byla přidána")
    else:
      print("Kniha se v knihovně již nachází")
  def removeKniha(self, kniha):
    if kniha not in self.seznamKnih:
      print("Kniha se v knihovně nenachází")
    else:
      self.seznamKnih.pop(self.seznamKnih.index(kniha))
      print("Kniha byla odebrána")
  def setNazev(self, nazev):
    self.nazev = nazev
  def setKapacitaKnih(self, kapacitaKnih:int):
    self.kapacitaKnih = kapacitaKnih
  def setSeznamKnih(self, SeznamKnih:list):
    self.seznamKnih = SeznamKnih
  def getAutor(self, autor):
    pocet = 0
    print(f"\n{autor} je autorem následujících knih v knihovně:")
    for kniha in self.seznamKnih:
      if kniha.getAutor().lower() == autor.lower():
        print(kniha)
  def __str__(self):
    temp = []
    for kniha in self.seznamKnih:
      temp.append(f"{kniha.getNazev()}, {kniha.getAutor()}, {kniha.getZanr()}, {kniha.getPocetStran()}")
    return(f"V Knihovně {self.nazev} je {len(self.seznamKnih)}/{self.kapacitaKnih} knih. Jedná se o: | {' | '.join(temp)} |")
```
