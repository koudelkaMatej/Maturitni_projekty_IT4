Vytvořte program, který vytvoří pole
teplot. Teploty budou zapsány výzvou
uživatele a uživatel takto zadá 4
teploty.
Teploty mohou obsahovat desetinné číslo
zaokrouhlené na 2 desetinné čárky a poté
vypíše průměrnou teplotu-pro každé roční
období.


```python
období = ["Jaro", "Léto", "Podzim" , "Zima"]
pole = []
w = 0
while (len(pole)<=3):
  x = round(float(input(f"Zadejteplotu za {období[w]}:")),2)
  w = w+1
  pole.append(x)
print(pole)

for x,pole in enumerate(pole):
  print(f"{období[x]}: {pole[x]}", "°C")
```

```python
# Každé třetí číslo ve vstupu se eliminuje
znamky = []
vahy = []
vysledek = []


k = 3


vstupRaw = input("Vstup: ")
docasna = vstupRaw.split(" ")
del docasna[k-1::k]
print(docasna)
if ("x" or "X") in docasna:
  index = docasna.index("x")
  docasna[index] = 10


for i in range(len(docasna)):
  if i % 2 == 0:  # Sudý index
      znamky.append(docasna[i])
  else:  # Lichý index
      vahy.append(docasna[i])
for x in range(len(znamky)):
  vysledek.append(float(znamky[x])*int(vahy[x]))


sumArr = sum(vysledek)
for y in range(len(vahy)):
  vahy[y] = int(vahy[y])
sumVah = sum(vahy)
print("Suma známek: ", sumArr, "| Suma vah:", sumVah)
print("\nPrůměr je: ", sumArr/sumVah)
```
