```python
import mysql.connector

# Procházení čísel uživatelů a databází od 1 do 40
for student_num in range(1, 41):
    try:
        # Připojení k databázi
        mydb = mysql.connector.connect(
            host="dbs.spskladno.cz",
            user=f"student{student_num}",
            password="spsnet",
            database=f"vyuka{student_num}"
        )

        mycursor = mydb.cursor()

        # Vypnutí kontrol cizích klíčů
        mycursor.execute("SET FOREIGN_KEY_CHECKS = 0")

        # Získání seznamu a smazání všech tabulek
        mycursor.execute("SHOW TABLES")
        for table in mycursor.fetchall():
            # Použití zpětných apostrofů kolem názvu tabulky
            mycursor.execute(f"DROP TABLE IF EXISTS `{table[0]}`")
            print(f"Tabulka {table[0]} v databázi vyuka{student_num} byla úspěšně smazána.")

        # Zapnutí kontrol cizích klíčů zpět
        mycursor.execute("SET FOREIGN_KEY_CHECKS = 1")

        mydb.commit()
        mycursor.close()
        mydb.close()

    except mysql.connector.Error as err:
        print(f"Chyba při práci s databází vyuka{student_num}: {err}")
```
