```python
def znam(date):
  day, month = map(int, date.split('.'))
  if (month ==3 and day >= 21) or (month == 4 and day <= 19):
    return "Beran"
  elif (month == 4 and day >= 20) or (month == 5 and day <= 20):
    return "Býk"
  elif (month == 5 and day >= 21) or (month == 6 and day <= 20):
    return "Blíženci"
  else:
    return "Neznám"

x = input()
print(znam(x))
```
