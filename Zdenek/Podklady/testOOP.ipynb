# 1
Vytvořte třídu Kniha, která obsahuje tyto atributy:

* Název
* Autor
* Žánr
* počet stran (přidejte omezení na vstupní hodnotu, která může být pouze INT)


a metody:
* get a set pro všechny atributy
* metodu ToString


```python
class Kniha:
    def __init__(self, nazev, autor, zanr, pocet_stran):
        self._nazev = nazev
        self._autor = autor
        self._zanr = zanr
        self.pocet_stran = pocet_stran

    @property
    def nazev(self):
        return self._nazev

    @nazev.setter
    def nazev(self, value):
        self._nazev = value

    @property
    def autor(self):
        return self._autor

    @autor.setter
    def autor(self, value):
        self._autor = value

    @property
    def zanr(self):
        return self._zanr

    @zanr.setter
    def zanr(self, value):
        self._zanr = value

    @property
    def pocet_stran(self):
        return self._pocet_stran

    @pocet_stran.setter
    def pocet_stran(self, value):
        if not isinstance(value, int):
            raise ValueError("Počet stran musí být celé číslo")
        self._pocet_stran = value

    def __str__(self):
        return f"Kniha: {self._nazev}, Autor: {self._autor}, Žánr: {self._zanr}, Počet stran: {self._pocet_stran}"

```

# 2
Vytvořte třídu Knihovna, která obsahuje tyto atributy:
* nazev
* seznamKnih
* kapacitaKnih

Vytvořte tyto metody:

* get a set pro všechny atributy
* metodu pro přepis všech knih do stringu
* ToString
* přidání a odebrání knihy z knihovny (lze přidat a odebrat pouze knihy, které se tam vejdou nebo v knihovně ještě nejsou + zpětná vazba uživateli)


```python
#TODO class Knihovna
class Knihovna:
    def __init__(self, nazev, kapacitaKnih):
        self._nazev = nazev
        self._seznamKnih = []
        self._kapacitaKnih = kapacitaKnih

    @property
    def nazev(self):
        return self._nazev

    @nazev.setter
    def nazev(self, value):
        self._nazev = value

    @property
    def seznamKnih(self):
        return self._seznamKnih

    @property
    def kapacitaKnih(self):
        return self._kapacitaKnih

    @kapacitaKnih.setter
    def kapacitaKnih(self, value):
        self._kapacitaKnih = value

    def pridatKnihu(self, kniha):
        if len(self._seznamKnih) >= self._kapacitaKnih:
            return "Knihovna je plná, nelze přidat další knihu."
        for k in self._seznamKnih:
            if k.nazev == kniha.nazev and k.autor == kniha.autor:
                return "Tato kniha je již v knihovně."
        self._seznamKnih.append(kniha)
        return f"Kniha '{kniha.nazev}' byla přidána do knihovny."


    def odebratKnihu(self, kniha):
        for k in self._seznamKnih:
            if k.nazev == kniha.nazev and k.autor == kniha.autor:
                self._seznamKnih.remove(k)
                return f"Kniha '{kniha.nazev}' byla odebrána z knihovny."
        return "Tato kniha není v knihovně."
    def vypisKnihyAutora(self, autor):
        knihy_od_autora = [k for k in self._seznamKnih if k.autor == autor]
        if knihy_od_autora:
            return "\n".join([str(k) for k in knihy_od_autora])
        return f"Žádné knihy od autora '{autor}' nebyly nalezeny."

    def __str__(self):
        knihy_str = ", ".join([f"{k.nazev} od {k.autor}" for k in self._seznamKnih]) if self._seznamKnih else "žádné knihy"
        return f"Knihovna '{self._nazev}' obsahuje {len(self._seznamKnih)}/{self._kapacitaKnih} knih: {knihy_str}"
```
