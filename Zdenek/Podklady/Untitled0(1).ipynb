Vtvořte aplikaci pro počítání kořenů u kvadratické rovnice
Vsutupem budou 4 proměnné.

Počítání kvadratické rovnice


```python
import math

def kvadraticka_funkce(a, b, c):
  D = (b)**2-4*a*c
  print("Diskrimninant je:",D)

  if D < 0:
    print("Nemá řešení")
  elif D==0:
    x = (-b/(2*a))
    print("Rovnice má pouze jedno řešení a to je:",x)
  else:
    x1 = ((-b+math.sqrt(D))/(2*a))
    x2 = ((-b-math.sqrt(D))/(2*a))
    print("Výsledky jsou","X1 je",x1,"a X2 je",x2)

kvadraticka_funkce(1,2,3)


```
