Vytvořte aplikaci, která se vás bude ve while cyklu ptát na jakou matematickou operaci chcete udělat. (sčítání, odčítání, násobvení nebo dělení). poté zadáte 2 čísla, které budete takto počítat. Na konci se Vás zeptá zda chcete zadat jiná čísla či skončíte pomocí stisku klávesy E.

```python

def scitani(a,b):
  scitani = a + b
  return scitani

def odcitani(a,b):
  odcitani = a-b
  return odcitani

def nasobveni(a,b):
  nasobveni = a*b
  return nasobveni

def deleni(a,b):
  deleni = a/b
  return deleni


while True:
  operace = str(input("Zdaejnte operaci(sčítání, odčítání, násobvení nebo dělení) "))
  if operace == "sčítání"or"+":
    prvni,druhy = int(input("Zadej prvni cislo ")), int(input(("Zadej druhé číslo ")))
    print("Součet čísel",prvni,"a ",druhy, "je: ",scitani(prvni,druhy))
  elif operace == "odčítání"or"-":
    prvni,druhy = int(input("Zadej prvni cislo ")), int(input(("Zadej druhé číslo ")))
    print("Rozdíl čísel",prvni,"a ",druhy, "je: ",odcitani(prvni,druhy))
  elif operace == "násobvení"or"*":
    prvni,druhy = int(input("Zadej prvni cislo ")), int(input(("Zadej druhé číslo ")))
    print("Součin čísel",prvni,"a ",druhy, "je: ",nasobveni(prvni,druhy))
  elif operace == "dělení"or"/"or"//"or"%":
    prvni,druhy = int(input("Zadej prvni cislo ")), int(input(("Zadej druhé číslo ")))
    print("Podíl čísel",prvni,"a ",druhy, "je: ",deleni(prvni,druhy))
  else:
    print("Seš idiot a zadal jsi to špatně")
  e = input("Pokud chcete ukončit operaci zadejte E ")
  if e == ("E")or("e"):
    break
```

Pomocí FOR cyklu vytvořte malou násobilku jednoho čísla

Výpis do konzole bude ve formátu:

1.5 x 0 = 0

2.5 x 1 = 5


```python
vstup = int(input("Zadejte číslo "))
for i in range(11):
  print("{} x {} = {}".format(i,vstup,vstup*i))
```


Vytvořte aplikaci, která se vás při startu zeptá na zadání čísla od 1 do 10. Pokud zadáte sudė číslo napíše se "Zadali jste sudé číslo.
Pokud napíšete liché číslo napíše se "Zadali jste liché číslo".
Pod tyto věty vytvořte pyramidu/ schody čísel. Např. když zadáte 3 tak pyramída vypadá
1
12
123
Poté vypíšete součet všech čísel v pyramidė.
Zjistěte datový typ proměnné sumy a vypište ho do konzole.

```python
vstup = int(input("Zadejte číslo od 1 do 10 "))
if vstup % 2 == 0:
  print("Zadali jste sudé číslo.")
else:
  print("Zadali jste liché číslo.")
soucet = 0
for i in range(vstup):
    for j in range(i+1):
        print("{}".format(j+1),end="")
        soucet += j
    print("")
print("Součet je: {}".format(soucet))
```
