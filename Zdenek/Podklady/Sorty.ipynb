Quicksort

```python
from random import randrange


def quicksort(array):
  if len(array) <= 1:
        return array
  pivot = array[randrange(len(array))]


  arrL = []
  arrM = []
  arrH = []


  for i in array:
        if i < pivot:
            arrL.append(i)
        elif i == pivot:
            arrM.append(i)
        else:
            arrH.append(i)


  return quicksort(arrL) + arrM + quicksort(arrH)


#pole = [3,1,5,7,9,8,4,2,0,6]
pole = []
for a in range(126666):
  pole.append(randrange(126666))
sort = quicksort(pole)
print(sort)
```

```python
# @title Výchozí text názvu
def quicksort(array, low, high):
    if low < high:
        pivot = partition(array, low, high)
        quicksort(array, low, pivot - 1)
        quicksort(array, pivot + 1, high)

def partition(array, low, high):
    pivot = array[high]
    i = low - 1

    for j in range(low, high):
        if array[j] <= pivot:
            i += 1
            array[i], array[j] = array[j], array[i]

    array[i + 1], array[high] = array[high], array[i + 1]
    return i + 1
array = []
quicksort(array, 0, len(array) - 1)
print(array)
```
