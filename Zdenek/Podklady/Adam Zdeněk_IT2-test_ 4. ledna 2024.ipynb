# Uživatel zadává název výrobku a jeho DPH
# Uživatel může také zadat, zda má na nějaký výrobek slevu, která činí 7% z původní hodnoty
# Hodnoty DPH mohou být pouze 15% a 21%
#Zadejte 3 výrobky, uveďte jejich DPH a zda lze uplatnit slevu
#Poté vypište všechny výrobky a jejich původní cenu a cenu po přidání DPH v Kč

```python
#TODO
nazev = [1]
DPH = [1]
sleva = [1]
cena = [1]
x= 1
while x <=1:
  nazev1 = input(f"Zadejte {x}. název: ")
  DPH1=int(input(f"Zadejte {x}. DPH: " ))
  if DPH1 != 15:
    if DPH1 !=21:
      DPH1=input(f"Jste kretén, DPH může být jen 15% a 21%. Zadejte znovu {x}. DPH: " )
  sleva1=int(input(f"Zadejte {x}. slevu: " ))
  cena1=int(input(f"Zadejte {x}. cenu: " ))
  nazev.append(nazev1)
  DPH.append(DPH1)
  sleva.append(sleva1)
  cena.append(cena1)
  x += 1

x= 1
while x < 2:

  krok1 = DPH[x]-sleva[x]
  krok2 = ((cena[x]*krok1)/100)
  krok4 = cena[x]+krok2
  print(f"{x}. výrobkem je: {nazev[x]} na který se vytahuje {DPH[x]}% DPH a k tomu je nabízena sleva {sleva[x]}%. Původní cena je: {cena[x]} a cena po přičtení DPH a odečtení slevy {krok4}")

  x += 1
```
