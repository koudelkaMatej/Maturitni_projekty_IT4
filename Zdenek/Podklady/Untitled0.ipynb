Postupně zadej libovolný počet čísel.
Ve chvíli, kdy jsou zadána všechna,
potvrd zadání zadáním slova "vykonej".
A následně dej na výběr, zda chceš čísla "sečíst" nebo pouze "vypsat" nebo určit, které číslo je liché a které sudé.
Použij funkce


```python

def sečti(čísla):
    return sum(čísla)

def vypiš(čísla):
    return čísla

def sudá_lichá(čísla):
    return {
        "sudá": [x for x in čísla if x % 2 == 0],
        "lichá": [x for x in čísla if x % 2 != 0]
    }

def hlavní():
    čísla = []
    while True:
        vstup = input("Zadej číslo nebo 'vykonej': ")
        if vstup.lower() == "vykonej":
            break
        try:
            čísla.append(int(vstup))
        except ValueError:
            print("Zadej prosím celé číslo nebo 'vykonej'.")

    print("Co chceš udělat? (sečíst / vypsat / sudé a liché)")
    volba = input("Zadej volbu: ").lower()

    if volba == "sečíst":
        print("Součet:", sečti(čísla))
    elif volba == "vypsat":
        print("Zadaná čísla:", vypiš(čísla))
    elif volba in ("sudé a liché", "liché a sudé"):
        výsledek = sudá_lichá(čísla)
        print("Sudá:", výsledek["sudá"])
        print("Lichá:", výsledek["lichá"])
    else:
        print("Neznámá volba.")

hlavní()
```
