Vytvořte Třídu Baťoh, která bude obsahovat atributy:

1. Předměty
2. Celková váha
3. Maximální váha

Dále obsahuje metody

1.   getPredmety, getCelkovaVaha, getMaximum
2.   setMaximum, addPredmet, addVaha, popVaha, removePredmet

Pokud přidáte do batohu větší váhu než je jeho maximální váha, předmět se nepřidá

```python
class Batoh:
    def __init__(self, max_vaha):
        self.predmety = []
        self.celkova_vaha = 0
        self.max_vaha = max_vaha

    def getPredmety(self):
        return self.predmety

    def getCelkovaVaha(self):
        return self.celkova_vaha

    def getMaximum(self):
        return self.max_vaha

    def setMaximum(self, max_vaha):
        self.max_vaha = max_vaha

    def addPredmet(self, predmet):
        if self.celkova_vaha + predmet.getVaha() <= self.max_vaha:
            self.predmety.append(predmet)
            self.celkova_vaha += predmet.getVaha()
        else:
            print("Přidání předmětu by překročilo maximální váhu batohu.")

    def removePredmet(self, predmet):
        if predmet in self.predmety:
            self.predmety.remove(predmet)
            self.celkova_vaha -= predmet.getVaha()
        else:
            print("Předmět není v batohu.")


```

Kontrola správného vzniku

```python
batoh = Batoh(10)
batoh.addPredmet(klacek)
batoh.addPredmet(mec)
print(batoh.getPredmety())
cihla = Predmet("cihla","těžkýýý",50)
batoh.addPredmet(cihla)
print(batoh.getPredmety())
batoh.removePredmet(klacek)
print(batoh.getPredmety())
```
