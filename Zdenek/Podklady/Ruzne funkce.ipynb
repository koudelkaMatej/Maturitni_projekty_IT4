Řešení kvadratické rovnice

```python
import math

def reseniKvadratickychRovnic(): # a * x^2 + b * x + c
  a = 1
  b = 1
  c = 1
  D = (b)**2-4*a*c


  if D < 0:
    print("Nemá řešení, protože diskriminant je:", D)
  elif D==0:
    x = (-b/(2*a))
    print("Rovnice má pouze jedno řešení a to je:",x)
  else:
    x1 = ((-b+math.sqrt(D))/(2*a))
    x2 = ((-b-math.sqrt(D))/(2*a))
    print("Výsledky jsou","X1 je",x1,"a X2 je",x2, "Diskriminant je:", D)
```

Euklidův algoritmus

```python
def eukliduvAlgoritmus(a,b):
  while b != 0:
        c = b
        b = a % b
        a = c
  return a
  a = a
  b = b

print("Euklid je:", eukliduvAlgoritmus(15, 20))
```

Hornerovo schéma

```python
def horneovoSchema(koeficienty, x):
  n = len(koeficienty) - 1
  hodnota = koeficienty[n]
  for i in range(n - 1, -1, -1):
    hodnota = hodnota * x + koeficienty[i]
  return hodnota
print(horneovoSchema([1, 2, 3], 2))
```

3n+1

```python
def triNJedna(n):
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1

print(triNJedna(3))
```

Faktoriál interativně

```python
def faktorial(n):
    if n == 0:
        return 1
    return n * faktorial(n - 1)

print(faktorial(5))

```

Faktoriál rekuzivne

```python
def faktorial(n):
    if n == 0:
        return 1
    else:
        return n * faktorial(n - 1)

print(faktorial(5))
```

Fibonaciho posloupnost iterativne

```python
def fibonacci(n):
    if n == 0 or n == 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(10))
```
