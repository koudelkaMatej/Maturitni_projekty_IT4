# 1
Vytvořte třídu Utulek, který bude obsahovat tyto vlastnosti:

*   Název
*   Maximální kapacita
*   SeznamZvířat

A jeho metody budou:

*  getNazev
*  getMaxKapacita
*  getSeznamZvirat
*  setNazev
*  setMaxKapacita
*  pridatZvire
*  odeberZvire
*  ToString







```python
class Utulek:
    def __init__(self, nazev, max_kapacita):
        self.nazev = nazev
        self.max_kapacita = max_kapacita
        self.seznamZvirat = []

    def getNazev(self):
        return self.nazev

    def getMaxKapacita(self):
        return self.max_kapacita

    def getSeznamZvirat(self):
        return self.seznamZvirat

    def setNazev(self, nazev):
        self.nazev = nazev

    def setMaxKapacita(self, max_kapacita):
        self.max_kapacita = max_kapacita

    def pridatZvire(self, zvire):
        if len(self.seznamZvirat) < self.max_kapacita:
            self.seznamZvirat.append(zvire)
        else:
            print("Utulek je plný, nelze přidat další zvíře.")

    def odeberZvire(self, zvire):
        if zvire in self.seznamZvirat:
            self.seznamZvirat.remove(zvire)
        else:
            print("Zvíře není v útulku.")

    def __str__(self):
        zvirata_str = ', '.join([str(z) for z in self.seznamZvirat])
        return f"Utulek {self.nazev}, Kapacita: {self.max_kapacita}, Zvířata: {zvirata_str}"
```

# 2
Vytvořte třídu Zvire, která bude obsahovat tyto vlasnosti
* Jméno
* Hmotnost

a metody budou pro get a set a ToString a metodu zarvi()

```python
class Zvire:
    def __init__(self, jmeno, hmotnost):
        self.jmeno = jmeno
        self.hmotnost = hmotnost

    def getJmeno(self):
        return self.jmeno

    def setJmeno(self, jmeno):
        self.jmeno = jmeno

    def getHmotnost(self):
        return self.hmotnost

    def setHmotnost(self, hmotnost):
        self.hmotnost = hmotnost

    def zarvi(self):
        return "Zvíře zarví!"

    def __str__(self):
        return f"Zvíře: {self.jmeno}, Hmotnost: {self.hmotnost} kg"
```
