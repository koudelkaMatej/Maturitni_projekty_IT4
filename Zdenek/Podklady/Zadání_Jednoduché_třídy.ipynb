upravte třídu Člověk tak, abyste ošetřili následující atributy:


1.   jméno:str

3.   telefon:int (maximálně 9 znaků)
4.   email:str (musí obsahovat @)
5. adresa:str
6. pojišťovna:str

Upravte metody, které budou moc měnit tyto atributy i s aplikovanými omezeními.




```python
class Clovek:
    def __init__(self, jmeno: str, prijmeni: str, telefon: int, email: str, adresa: str, pojistovna: str):
        self.set_jmeno(jmeno)
        self.set_prijmeni(prijmeni)
        self.set_telefon(telefon)
        self.set_email(email)
        self.set_adresa(adresa)
        self.set_pojistovna(pojistovna)

    ## Gettery
    def get_jmeno(self):
        return self.jmeno

    def get_prijmeni(self):
        return self.prijmeni

    def get_telefon(self):
        return self.telefon

    def get_email(self):
        return self.email

    def get_adresa(self):
        return self.adresa

    def get_pojistovna(self):
        return self.pojistovna

    ## Settery s kontrolami
    def set_jmeno(self, jmeno):
        self.jmeno = jmeno

    def set_prijmeni(self, prijmeni):
        self.prijmeni = prijmeni

    def set_telefon(self, telefon):
        if len(str(telefon)) <= 9:
            self.telefon = telefon
        else:
            raise ValueError("Telefonní číslo nesmí být delší než 9 číslic.")

    def set_email(self, email):
        if "@" in email:
            self.email = email
        else:
            raise ValueError("E-mail musí obsahovat znak '@'.")

    def set_adresa(self, adresa):
        self.adresa = adresa

    def set_pojistovna(self, pojistovna):
        self.pojistovna = pojistovna

    def vypis(self):
        return f"Jméno: {self.jmeno}, Příjmení: {self.prijmeni}, Telefon: {self.telefon}, Email: {self.email}, Adresa: {self.adresa}, Pojišťovna: {self.pojistovna}"


```

Vytvořte třídu pojištovna, která bude mít atributy:


1.   název:str
2.   zkratkaPojištovny:str (maximálně 3 znaky)
3. sídlo:str

Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše o této pojišťovně.




```python
class Pojistovna:
    def __init__(self, nazev, zkratka, sidlo):
        self.nazev = nazev
        self.zkratka = zkratka
        self.sidlo = sidlo

    def set_nazev(self, novy_nazev):
        self.nazev = novy_nazev

    def set_zkratka(self, nova_zkratka):
        if len(nova_zkratka) <= 3:
            self.zkratka = nova_zkratka
        else:
            print("Chyba: Zkratka pojišťovny může mít maximálně 3 znaky.")

    def set_sidlo(self, nove_sidlo):
        self.sidlo = nove_sidlo

    def get_nazev(self):
        return self.nazev

    def get_zkratka(self):
        return self.zkratka

    def get_sidlo(self):
        return self.sidlo

    def vypis_informace(self):
        print(f"Název pojišťovny: {self.nazev}")
        print(f"Zkratka pojišťovny: {self.zkratka}")
        print(f"Sídlo pojišťovny: {self.sidlo}")

```

```python
#Zde vytvořte alespoň 2 pojišťovny a zkuste si všechny metody na jedné z nich
pojistovna1 = Pojistovna("Pojišťovna A.S.", "PAS", "Praha")
pojistovna2 = Pojistovna("Univerzální Pojišťovna", "UP", "Brno")
pojistovna3 = Pojistovna("Zdravotní Pojišťovna", "ZP", "Ostrava")
pojistovna4 = Pojistovna("Cestovní Pojišťovna", "CP", "Plzeň")
pojistovna5 = Pojistovna("Auto Pojišťovna", "AP", "Liberec")

# Testování metod na pojistovně 1
# Výpis původních informací
pojistovna1.vypis_informace()

# Změna názvu, zkratky a sídla
pojistovna1.set_nazev("Nová Pojišťovna A.S.")
pojistovna1.set_zkratka("NPA")
pojistovna1.set_sidlo("Karlovy Vary")

# Výpis aktualizovaných informací
pojistovna1.vypis_informace()
```

#4. Zadání

Upravte třídu Člověk tak, aby atribut pojišťovna byl vložen jako třída Pojišťovna


1.   jméno:str
2.   příjmení:str
3.   telefon:int (maximálně 9 znaků)
4.   email:str (musí obsahovat @)
5. adresa:str
6. pojišťovna:class(Pojišťovna)

Upravte metodu, která vám vypíše vše o této osobě tak, aby Vám také vypsala všechny informace o dané pojišťovně




```python
class Clovek:
    def __init__(self, jmeno: str, prijmeni: str, telefon: int, email: str, adresa: str, pojistovna: Pojistovna):
        self.set_jmeno(jmeno)
        self.set_prijmeni(prijmeni)
        self.set_telefon(telefon)
        self.set_email(email)
        self.set_adresa(adresa)
        self.set_pojistovna(pojistovna)

    ## Gettery
    def get_jmeno(self):
        return self.jmeno

    def get_prijmeni(self):
        return self.prijmeni

    def get_telefon(self):
        return self.telefon

    def get_email(self):
        return self.email

    def get_adresa(self):
        return self.adresa

    def get_pojistovna(self):
        return self.pojistovna

    ## Settery s kontrolami
    def set_jmeno(self, jmeno):
        self.jmeno = jmeno

    def set_prijmeni(self, prijmeni):
        self.prijmeni = prijmeni

    def set_telefon(self, telefon):
        if len(str(telefon)) <= 9:
            self.telefon = telefon
        else:
            raise ValueError("Telefonní číslo nesmí být delší než 9 číslic.")

    def set_email(self, email):
        if "@" in email:
            self.email = email
        else:
            raise ValueError("E-mail musí obsahovat znak '@'.")

    def set_adresa(self, adresa):
        self.adresa = adresa

    def set_pojistovna(self, pojistovna):
        if isinstance(pojistovna, Pojistovna):
            self.pojistovna = pojistovna
        else:
            raise ValueError("Pojišťovna musí být instancí třídy Pojistovna.")

    def vypis(self):
        pojistovna_info = f"Název pojišťovny: {self.pojistovna.get_nazev()}, Zkratka: {self.pojistovna.get_zkratka()}, Sídlo: {self.pojistovna.get_sidlo()}"
        return f"Jméno: {self.jmeno}, Příjmení: {self.prijmeni}, Telefon: {self.telefon}, Email: {self.email}, Adresa: {self.adresa}, {pojistovna_info}"
```

```python
pojistovna1 = Pojistovna("Česká pojišťovna", "ČP", "Praha")
pojistovna2 = Pojistovna("Allianz", "ALZ", "Brno")

osoba1 = Clovek("Jan", "Novák", 123456789, "jan.novak@email.cz", "Praha 1", pojistovna1)
osoba2 = Clovek("Eva", "Svobodová", 987654321, "eva.svoboda@email.cz", "Brno 2", pojistovna2)
# Výpis informací o osobě
print(osoba1.vypis())

# Změna telefonního čísla (příklad)
try:
    osoba1.set_telefon(111222333)
except ValueError as e:
    print(e)

# Výpis nového telefonního čísla
print(osoba1.get_telefon())

# Změna e-mailu (příklad)
try:
    osoba1.set_email("novy.email@email.cz")
except ValueError as e:
    print(e)

# Výpis nového e-mailu
print(osoba1.get_email())
```

Vytvořte třídu Pocasi, která bude mít atributy:

1.   vítr:float
2.   povetrnostniPodminky:str
3.   teplota:float
4.   vlhkost:float

Vytvořte metody, které budou moc získávat a měnit tyto atributy. Také vytvořte metodu, která vám vypíše vše o této osobě


```python
class Pocasi:
    #TODO
    def __init__(self, vitr: float, povetrnostniPodminky: str, teplota: float, vlhkost: float):
        self.vitr = vitr
        self.povetrnostniPodminky = povetrnostniPodminky
        self.teplota = teplota
        self.vlhkost = vlhkost

    def get_vitr(self):
        return self.vitr

    def set_vitr(self, vitr: float):
        self.vitr = vitr

    def get_povetrnostniPodminky(self):
        return self.povetrnostniPodminky

    def set_povetrnostniPodminky(self, podminky: str):
        self.povetrnostniPodminky = podminky

    def get_teplota(self):
        return self.teplota

    def set_teplota(self, teplota: float):
        self.teplota = teplota

    def get_vlhkost(self):
        return self.vlhkost

    def set_vlhkost(self, vlhkost: float):
        self.vlhkost = vlhkost

    def vypis_info(self):
        info = (f"Vítr: {self.vitr} km/h, "
                f"Povětrnostní podmínky: {self.povetrnostniPodminky}, "
                f"Teplota: {self.teplota}°C, "
                f"Vlhkost: {self.vlhkost}%")
        print(info)
```

```python
#Zde vytvořte alespoň 2 záznamy počasí a zkuste si všechny metody na jedné z nich
# Vytvoření dvou instancí třídy Pocasi
pocasi1 = Pocasi(vitr=15.0, povetrnostniPodminky="Jasno", teplota=25.3, vlhkost=45.0)
pocasi2 = Pocasi(vitr=5.0, povetrnostniPodminky="Oblačno", teplota=18.5, vlhkost=60.0)

# Výpis počátečních hodnot
print("Původní hodnoty pocasi1:")
pocasi1.vypis_info()

# Změna hodnot pomocí setter metod
pocasi1.set_vitr(20.0)
pocasi1.set_povetrnostniPodminky("Déšť")
pocasi1.set_teplota(22.0)
pocasi1.set_vlhkost(70.0)

# Výpis upravených hodnot
print("\nUpravené hodnoty pocasi1:")
pocasi1.vypis_info()

# Získání konkrétních hodnot pomocí getter metod
print("\nKonkrétní hodnoty z pocasi1:")
print(f"Vítr: {pocasi1.get_vitr()} km/h")
print(f"Povětrnostní podmínky: {pocasi1.get_povetrnostniPodminky()}")
print(f"Teplota: {pocasi1.get_teplota()}°C")
print(f"Vlhkost: {pocasi1.get_vlhkost()}%")
```
