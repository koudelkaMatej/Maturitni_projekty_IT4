Kalkulačka: Vytvořte program, který bude fungovat jako kalkulačka. Program by měl umět provádět základní aritmetické operace (sčítání, odčítání, násobení, dělení), ale také by měl umět provádět operace s desetinnými čísly a zlomky.

```python
import math

def scitani(a, b):
    return a + b

def odcitani(a, b):
    return a - b

def nasobeni(a, b):
    return a * b

def deleni(a, b):
    return a / b

def mocnina(a, b):
    return a ** b

def odmocnina(a):
    return math.sqrt(a)

def logaritmus(a, b):
    return math.log(a, b)

print("Kalkulačka")
a = float(input("Zadej první číslo: "))
b = float(input("Zadej druhé číslo: "))
print("Zadej operaci (+, -, *, /, ^, √, ln): ")
operace = input()

if operace == "+":
    print(a, "+", b, "=", scitani(a, b))
elif operace == "-":
    print(a, "-", b, "=", odcitani(a, b))
elif operace == "*":
    print(a, "*", b, "=", nasobeni(a, b))
elif operace == "/":
    print(a, "/", b, "=", deleni(a, b))
elif operace == "^":
    print(a, "^", b, "=", mocnina(a, b))
elif operace == "√":
    print("√", a, "=", odmocnina(a))
elif operace == "ln":
    print("ln", a, "=", logaritmus(a, math.e))
else:
    print("Neznámá operace")
```

Vytvořte program v Pythonu, který bude provádět následující úkoly:

Načte od uživatele číslo n.
Vypočítá a vypíše Fibonacciho čísla do n.

```python
def fibonacci(n):
    if n == 0 or n == 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)


cislo = int(input("Zadejte číslo n: "))

for i in range(cislo):
    print(fibonacci(i))
```

Vytvořte program v Pythonu, který bude provádět následující úkoly:

Načte od uživatele číslo n.
Vypočítá a vypíše faktoriál čísla n.

```python
def faktorial(n):
    if n == 0:
        return 1
    else:
        return n * faktorial(n - 1)


cislo = int(input("Zadejte číslo n: "))

print(faktorial(cislo))
```

Vytvořte program v Pythonu, který bude provádět následující úkoly:

Načte od uživatele číslo n.
Vypíše všechna prvočísla menší nebo rovná n.

```python
def je_prvocislo(cislo):
    for i in range(2, int(cislo ** 0.5) + 1):
        if cislo % i == 0:
            return False
    return True


cislo = int(input("Zadejte číslo n: "))

for i in range(2, cislo + 1):
    if je_prvocislo(i):
        print(i)
```

Vytvořte program v Pythonu, který bude provádět následující úkoly:

Načte od uživatele text.
Vyhledá v textu všechna slova, která se vyskytují více než dvakrát.

```python
def najdi_slova(text):
    slova = text.split()
    slovnik = {}
    for slovo in slova:
        if slovo in slovnik:
            slovnik[slovo] += 1
        else:
            slovnik[slovo] = 1
    return [slovo for slovo, pocet in slovnik.items() if pocet > 2]


text = input("Zadejte text: ")

slova = najdi_slova(text)

for slovo in slova:
    print(slovo)
```

Vytvořte program v Pythonu, který bude provádět následující úkoly:

Načte od uživatele číslo n.
Vykreslí na obrazovku n-úhelník.

```python
import turtle

def draw_n_angle(n, size):
    for i in range(n):
        turtle.forward(size)
        turtle.left(360 / n)

def main():
    n = int(input("Zadej počet úhlů: "))
    size = float(input("Zadej velikost úhlu: "))
    turtle.setup(width=400, height=400)
    turtle.speed(0)
    draw_n_angle(n, size)

if __name__ == "__main__":
    main()
```
